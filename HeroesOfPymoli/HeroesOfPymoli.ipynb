```python
#Dependencies and setup
import pandas as pd
import os
import csv


# Import csv file
filepath = "purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(filepath)
```

```python
#Display the total number of players
total_players = len(purchase_data["SN"].unique())
total_players_df = pd.DataFrame({"Total Players":[total_players]})
total_players_df
```

```python
# Purchasing Analysis

# Number of Unique Items
item_id = (purchase_data['Item ID'].nunique())

# Average Purchase Price
average_price = purchase_data["Price"].mean()
average_price = "${0:,.2f}".format(average_price)

# Total Number of Purchases
num_of_purchases = purchase_data.shape[0]

# Total Revenue
total_revenue = purchase_data["Price"].sum()
total_revenue = "${0:,.2f}".format(total_revenue)


purchasing_analysis ={"Number of Unique Items": [item_id],
                         "Average Price": [average_price],
                         "Number of Purchases": [num_of_purchases],
                         "Total Revenue": [total_revenue]}

purchasing_analysis_df = pd.DataFrame(purchasing_analysis, columns = ["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"])
purchasing_analysis_df
```

```python
# Gender Demographics

# Percentage and Count of Male Players
male_count =(len(purchase_data.query('Gender == "Male"').SN.unique()))
male_percentage = format((male_count/total_players)*100, ".2f")
print(male_percentage)

# Percentage and Count of Female Players
female_count =(len(purchase_data.query('Gender == "Female"').SN.unique()))
female_percentage = format((female_count/total_players)*100, ".2f")
print(female_percentage)

# Percentage and Count of Other/Non-Disclosed Players
other_nondisclosed_count = (len(purchase_data.query('Gender == "Other / Non-Disclosed"').SN.unique()))
other_nondisclosed_percentage = format((other_nondisclosed_count/total_players)*100, ".2f")
print(other_nondisclosed_percentage
     )
gender_demographics = {'Gender':['Male', 'Female','Other / Non-Disclosed'],
                       'Total Count':[male_count, female_count, other_nondisclosed_count],
                       'Percentage':[male_percentage, female_percentage, other_nondisclosed_percentage]}
gender_demographics_df = pd.DataFrame(gender_demographics, columns = ['Gender', 'Total Count', 'Percentage'])
gender_demographics_df
```

```python
# Figure out minimum and maximum ages of players

min_age = (purchase_data["Age"].min()) 
max_age = (purchase_data["Age"].max())
#print(min_age, max_age)

# Create bins
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 1000]

bin_labels = ["< 10", "10 - 14", "15 - 19", "20 - 24", "25 - 29",
              "30 - 34", "35 - 39", "40+"]

# Slice the data and place it into the bins
purchase_data["Age Demographic Breakdown"]= pd.cut(purchase_data["Age"], age_bins, 
                                                   labels = bin_labels, include_lowest = True)

# Add new columns
age_counts = purchase_data.groupby("Age Demographic Breakdown").size().reset_index(name="Total Count")
age_counts["Percentage of Players"] = age_counts[["Total Count"]]/total_players * 100

# Remove index

age_counts
```

```python
# Top Spender Trends

# Top 5 spenders values
# Used aggregate to run count, mean, and sum--for count, average, and total value
individual_totals = purchase_data[["SN", "Price"]].groupby("SN").agg(["count","mean","sum"])

#Renaming aggregate columns
individual_totals.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value"] 

#Creates data frame and sorts the top 5 values
top_spender_data = individual_totals.sort_values("Total Purchase Value", ascending = False).head()
top_spender_data
```

```python
# Most Popular Items

# Used aggregate to run count, mean, and sum--for count, average, and total value

item_totals = purchase_data[["Item ID", "Item Name", "Price"]].groupby(["Item ID", "Item Name"]).agg(["count", "max", "sum"])
item_totals

#Renaming aggregate columns
item_totals.columns = ["Purchase Count", "Item Price", "Total Purchase Value"]
item_totals

#Create new dataframe
most_popular_items = item_totals.sort_values("Purchase Count", ascending  = False).head()
most_popular_items
```

```python
# Most Profitable Items

# Used aggregate to run count, mean, and sum--for count, average, and total value

item_totals_profit = purchase_data[["Item ID", "Item Name", "Price"]].groupby(["Item ID", "Item Name"]).agg(["count", "max", "sum"])
item_totals_profit

#Renaming aggregate columns
item_totals_profit.columns = ["Purchase Count", "Item Price", "Total Purchase Value"]
item_totals_profit

#Create new dataframe
most_profitable_items = item_totals.sort_values("Total Purchase Value", ascending  = False).head()
most_profitable_items
```
